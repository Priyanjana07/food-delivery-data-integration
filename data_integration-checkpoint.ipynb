```python
orders_df = pd.read_csv("orders.csv")
orders_df.head()
users_df = pd.read_json("users.json")
users_df.head()
with open(
   r"C:\Users\priya\Downloads\restaurants.sql", "r"
) as f:
    sql_script = f.read()


orders_users_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)

final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)

final_df.head()
```

```python

revenue_by_combo = final_df.groupby(["membership", "cuisine"])["total_amount"].sum()


highest_revenue_combo = revenue_by_combo.idxmax()
highest_revenue_amount = revenue_by_combo.max()

print("Highest revenue combination:", highest_revenue_combo)
print("Revenue amount:", highest_revenue_amount)
```

```python

revenue_by_quarter = final_df.groupby("quarter")["total_amount"].sum()


highest_revenue_quarter = revenue_by_quarter.idxmax()
highest_revenue_amount = revenue_by_quarter.max()

print("Quarter with highest revenue:", highest_revenue_quarter)
print("Revenue amount:", highest_revenue_amount)
```

```python

restaurant_stats = final_df.groupby("restaurant_name_y").agg(
    orders=("order_id", "count"),
    aov=("total_amount", "mean")
)


small_restaurants = restaurant_stats[restaurant_stats["orders"] < 20]


highest_aov_restaurant = small_restaurants["aov"].idxmax()
highest_aov_value = small_restaurants["aov"].max()

print("Restaurant with highest AOV (<20 orders):", highest_aov_restaurant)
print("Average Order Value:", round(highest_aov_value, 2))
```
